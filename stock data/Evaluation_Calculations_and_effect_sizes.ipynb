```python
#Experiment 1
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score

# File paths
file_paths = {
    "arima": "arima_index_price_predictions.csv",
    "lgbm_finbert": "lgbm_index_actual_vs_predicted_finbert.csv",
    "lgbm_no_sentiment": "lgbm_index_actual_vs_predicted_no_sentiment.csv",
    "lgbm_roberta": "lgbm_index_actual_vs_predicted_roberta.csv",
    "lstm": "lstm_predicted_vs_actual_index.csv",
    "lstm_finbert": "lstm_with_finbert_index.csv",
    "lstm_roberta": "lstm_with_roberta_index.csv"
}

# Column mapping for each file (actual vs predicted)
column_map = {
    "arima": ("Actual", "Predicted"),
    "lgbm_finbert": ("Index_Level", "Predicted_Index_Level"),
    "lgbm_no_sentiment": ("Index_Level", "Predicted_Index_Level"),
    "lgbm_roberta": ("Index_Level", "Predicted_Index_Level"),
    "lstm": ("Index_Level", "Predicted_Index_Level"),
    "lstm_finbert": ("Index_Level", "Predicted_Index_Level"),
    "lstm_roberta": ("Index_Level", "Predicted_Index_Level"),
}

# Function to compute metrics
def calculate_metrics(df, actual_col, predicted_col):
    mae = mean_absolute_error(df[actual_col], df[predicted_col])
    mape = mean_absolute_percentage_error(df[actual_col], df[predicted_col])
    r2 = r2_score(df[actual_col], df[predicted_col])
    return mae, mape, r2

# Process all files and compute metrics
results = []
for name, path in file_paths.items():
    df = pd.read_csv(path)
    actual_col, predicted_col = column_map[name]
    mae, mape, r2 = calculate_metrics(df, actual_col, predicted_col)
    results.append({
        "Model": name,
        "MAE": mae,
        "MAPE": mape,
        "R2": r2
    })

# Create results DataFrame and print
results_df = pd.DataFrame(results)
print(results_df)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro

# Load CSVs
df_lstm = pd.read_csv("lstm_predicted_vs_actual_index.csv")
df_lgbm = pd.read_csv("lgbm_index_actual_vs_predicted_no_sentiment.csv")
df_arima = pd.read_csv("arima_index_price_predictions.csv")

# Compute absolute errors
lstm_errors = np.abs(df_lstm["Index_Level"] - df_lstm["Predicted_Index_Level"])
lgbm_errors = np.abs(df_lgbm["Index_Level"] - df_lgbm["Predicted_Index_Level"])
arima_errors = np.abs(df_arima["Actual"] - df_arima["Predicted"])

# Run Shapiro-Wilk test for normality
def print_shapiro(label, errors):
    stat, p = shapiro(errors)
    print(f"{label} - Shapiro-Wilk W={stat:.4f}, p={p:.4e}")

print_shapiro("LSTM", lstm_errors)
print_shapiro("LGBM", lgbm_errors)
print_shapiro("ARIMA", arima_errors)

# Plotting
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
sns.histplot(lstm_errors, kde=True)
plt.title("LSTM Baseline: Absolute Error Distribution")
plt.xlabel("Absolute Error")

plt.subplot(1, 3, 2)
sns.histplot(lgbm_errors, kde=True)
plt.title("LGBM Baseline: Absolute Error Distribution")
plt.xlabel("Absolute Error")

plt.subplot(1, 3, 3)
sns.histplot(arima_errors, kde=True)
plt.title("ARIMA: Absolute Error Distribution")
plt.xlabel("Absolute Error")

plt.tight_layout()
plt.savefig("error_distributions.png", dpi = 300)
plt.show()

```

```python
import pandas as pd
import numpy as np
import itertools

# Define file paths
files = {
    "lstm_baseline": "lstm_predicted_vs_actual_index.csv",
    "lstm_finbert": "lstm_with_finbert_index.csv",
    "lstm_roberta": "lstm_with_roberta_index.csv",
    "lgbm_baseline": "lgbm_index_actual_vs_predicted_no_sentiment.csv",
    "lgbm_finbert": "lgbm_index_actual_vs_predicted_finbert.csv",
    "lgbm_roberta": "lgbm_index_actual_vs_predicted_roberta.csv",
    "arima_baseline": "arima_index_price_predictions.csv"
}

# Cliff's delta function
def cliffs_delta(x, y):
    n_x = len(x)
    n_y = len(y)
    greater = sum(1 for xi, yj in itertools.product(x, y) if xi > yj)
    lesser = sum(1 for xi, yj in itertools.product(x, y) if xi < yj)
    return (greater - lesser) / (n_x * n_y)

# Load all errors into a dictionary
errors = {}
for name, path in files.items():
    df = pd.read_csv(path)
    actual = df["Index_Level"] if "Index_Level" in df.columns else df["Actual"]
    predicted = df["Predicted_Index_Level"] if "Predicted_Index_Level" in df.columns else df["Predicted"]
    errors[name] = np.abs(actual - predicted)

# Define model comparisons
comparisons = [
    ("lstm_baseline", "lstm_finbert"),
    ("lstm_baseline", "lstm_roberta"),
    ("lgbm_baseline", "lgbm_finbert"),
    ("lgbm_baseline", "lgbm_roberta"),
    ("arima_baseline", "lstm_baseline"),
    ("arima_baseline", "lgbm_baseline")
]

# Compute Cliff's Delta for each comparison
results = []
for base, model in comparisons:
    delta = cliffs_delta(errors[base], errors[model])
    results.append({
        "Baseline": base,
        "Model": model,
        "Cliff's Delta": round(delta, 3)
    })

# Create and display the results table
results_df = pd.DataFrame(results)
print(results_df)
```
